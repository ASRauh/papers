```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-paper')

#Load notag Rgs from file
NaCl_notag = [150,1000]
NaCl = [50,100,200,300,400,500]
Rg = []
Rg_avg = []

Rg_fornorm = []
Rg_avg_fornorm = []

#Calculate Rg average from .xvg files and append to Rg_avg
for i in range(len(NaCl_notag)):
    Rg.append(np.genfromtxt('Rg_gyrate_' + str(NaCl_notag[i]) + 'mM.xvg',skip_header=27,skip_footer=0,usecols=[1],unpack=True))
    Rg_avg.append(np.average(Rg[i]))

#Calculate Rg average from .xvg files only for data points shared with His-SUMO construct
for i in range(len(NaCl)):
    Rg_fornorm.append(np.genfromtxt('Rg_gyrate_' + str(NaCl[i]) + 'mM.xvg',skip_header=27,skip_footer=0,usecols=[1],unpack=True))
    Rg_avg_fornorm.append(np.average(Rg_fornorm[i]))
    
#Plot Rg averages
fig= plt.figure(figsize=(6,5))
axes= fig.add_axes([1,1,1,1])

plt.plot(NaCl_notag, Rg_avg, 'o-', linewidth=2.5, markersize=12, color='black')


#plt.ylim(2.8, 3.3)
#plt.xlim(40, 510)


plt.xlabel('[NaCl] (mM)', fontsize=25)
plt.ylabel('R$_g$ (nm)', fontsize=25)

plt.xticks(np.arange(100, 1001, step=200),fontsize=20)
plt.yticks(np.arange(2.8, 4.2, step=0.2),fontsize=20) 

plt.tick_params(colors='black', direction='out', length=7)

fig.savefig("/lindorffgrp-isilon/thomasen/hnRNPA1/ionicstrength_notag_productionrun/figures/Rgcoords_vs_NaCl.png", bbox_inches='tight', pad_inches=1)
```

```python
#Normalize Rgs to average of Rgs across salt concentrations
Rg_avg_scaled=[]

for i in range(len(Rg_avg)):
    scaleRg=Rg_avg[i]/np.average(Rg_avg_fornorm)
    Rg_avg_scaled.append(scaleRg)

fig= plt.figure(figsize=(6,5))
axes= fig.add_axes([1,1,1,1])

plt.plot(NaCl_notag, Rg_avg_scaled, 'o-', linewidth=2.5, markersize=12, color='black')


#plt.ylim(2.8, 3.3)
#plt.xlim(40, 510)

plt.xlabel('[NaCl] (mM)', fontsize=25)
plt.ylabel('Normalized R$_g$', fontsize=25)

#plt.xticks(np.arange(100, 501, step=100),fontsize=20)
#plt.yticks(np.arange(2.8, 3.35, step=0.1),fontsize=20) 

plt.tick_params(colors='black', direction='out', length=7)
```

```python
NaCl = [50,100,200,300,400,500]
Rg_hissumotag = []
Rg_hissumotag_avg = []

di="/lindorffgrp-isilon/thomasen/hnRNPA1/hnRNPA1_ionicstrength_2/data/"

#Calculate Rg average from .xvg files
for i in range(len(NaCl)):
    Rg_hissumotag.append(np.genfromtxt(di + 'Rg_minusstart_hnRNPA1_' + str(NaCl[i]) + 'mM_theta1.07.xvg',skip_header=27,skip_footer=0,usecols=[1],unpack=True))
    Rg_hissumotag_avg.append(np.average(Rg_hissumotag[i]))
    
#Scale Rgs by average of Rgs across salt concentrations
Rg_hissumotag_avg_scaled=[]

for i in range(len(Rg_hissumotag_avg)):
    scaleRg_hissumotag=Rg_hissumotag_avg[i]/np.average(Rg_hissumotag_avg)
    Rg_hissumotag_avg_scaled.append(scaleRg_hissumotag)
    
fig= plt.figure(figsize=(6,5))
axes= fig.add_axes([1,1,1,1])

plt.plot(NaCl, Rg_hissumotag_avg_scaled, 'o-', linewidth=2.5, markersize=12, color='black')


#plt.ylim(2.8, 3.3)
#plt.xlim(40, 510)

plt.xlabel('[NaCl] (mM)', fontsize=25)
plt.ylabel('Normalized R$_g$', fontsize=25)

#plt.xticks(np.arange(100, 501, step=100),fontsize=20)
#plt.yticks(np.arange(2.8, 3.35, step=0.1),fontsize=20) 

plt.tick_params(colors='black', direction='out', length=7)
```

```python
fig= plt.figure(figsize=(6,5))
axes= fig.add_axes([1,1,1,1])

plt.plot(NaCl_notag, Rg_avg_scaled, 'o-', linewidth=2.5, markersize=12, label="no tag")
plt.plot(NaCl, Rg_hissumotag_avg_scaled, 'o-', linewidth=2.5, markersize=12,label="with His-SUMO tag")


plt.ylim(0.87, 1.42)
plt.xlim(-10, 1010)

plt.legend(fontsize=15)
plt.xlabel('[NaCl] (mM)', fontsize=25)
plt.ylabel('Normalized R$_g$', fontsize=25)

plt.xticks(np.arange(0, 1010, step=200),fontsize=20)
plt.yticks(np.arange(0.90, 1.42, step=0.1),fontsize=20) 

plt.tick_params(colors='black', direction='out', length=7)

fig.savefig("/lindorffgrp-isilon/thomasen/hnRNPA1/ionicstrength_notag_productionrun/figures/Rgcoords_normalizedtoavgRg_vs_NaCl_notag_and_tag", bbox_inches='tight', pad_inches=1)
```

```python
#Normalize Rg to nr of residues

notag_resnr=433-119
tag_resnr=433

Rg_notag_avg_scaled_res=[]
Rg_tag_avg_scaled_res=[]

#Normalize notag
for i in range(len(Rg_avg)):
    scaleRg=Rg_avg[i]/notag_resnr
    Rg_notag_avg_scaled_res.append(scaleRg)

#Normalize tagged
for i in range(len(Rg_hissumotag_avg)):
    scaleRg=Rg_hissumotag_avg[i]/tag_resnr
    Rg_tag_avg_scaled_res.append(scaleRg)

print(Rg_notag_avg_scaled_res)
print(Rg_tag_avg_scaled_res)
```

```python
fig= plt.figure(figsize=(6,5))
axes= fig.add_axes([1,1,1,1])

plt.plot(NaCl_notag, Rg_notag_avg_scaled_res, 'o-', linewidth=2.5, markersize=12, label="no tag")
plt.plot(NaCl, Rg_tag_avg_scaled_res, 'o-', linewidth=2.5, markersize=12,label="with His-SUMO tag")


#plt.ylim(0.87, 1.35)
plt.xlim(-10, 1010)

plt.legend(fontsize=15)
plt.xlabel('[NaCl] (mM)', fontsize=25)
plt.ylabel('Normalized R$_g$', fontsize=25)

plt.xticks(np.arange(0, 1010, step=200),fontsize=20)
#plt.yticks(np.arange(0.90, 1.36, step=0.1),fontsize=20) 

plt.tick_params(colors='black', direction='out', length=7)

#fig.savefig("", bbox_inches='tight', pad_inches=1)
```
