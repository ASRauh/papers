```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-paper')
```

# Lambda scan

```python
#lambda called theta here

def plot_chi2r_vs_lambda(pepsiparams):
    
    #mean or opt pepsiparams
    if pepsiparams == "mean":
        colnum=1
    elif pepsiparams == "opt":
        colnum=2
        
    #get chi2r from summary files
    thetas = ['1.00', '1.04', '1.06', '1.07', '1.08', '1.09', '1.10', '1.12']
    chi2rlist = []
    
    for theta in thetas:
        dir = '/lindorffgrp-isilon/thomasen/hnRNPA1/PWrescaling_notag_productionrun_150mM/theta' + str(theta) + '/stage_minusstart'
        chi2r = np.genfromtxt('%s/summary_0_file.dat' % dir, skip_header=1, usecols=(colnum))
        chi2rlist.append(chi2r)
    
    print(chi2rlist)
    
    #plot    
    fig=plt.figure(figsize=(6,5))
    axes=fig.add_axes([1,1,1,1])
    
    plt.plot(thetas, chi2rlist, 'o-', linewidth=2.5, markersize=12)
    
    plt.title('Pepsi params are ' + str(pepsiparams), fontsize=15)
    
    plt.xlabel('lambda', fontsize=20)
    plt.ylabel('chi2r', fontsize=20)
    
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20) 

    plt.tick_params(colors='black', direction='out', length=7)
    
    fig.savefig('/lindorffgrp-isilon/thomasen/hnRNPA1/PWrescaling_notag_productionrun_150mM/figures/chi2r_vs_lambda_pepsiparams' + str(pepsiparams) + '.png', bbox_inches='tight', pad_inches=1)
```

```python
plot_chi2r_vs_lambda("opt")
plot_chi2r_vs_lambda("mean")
```

## SI Plot

```python
#get chi2r from summary files
thetas = ['1.0', '1.02', '1.04', '1.05', '1.06', '1.07', '1.08', '1.10']
thetas_float_withtag = [1.00, 1.02, 1.04, 1.05, 1.06, 1.07, 1.08, 1.10]

chi2rlist_withtag = []
    
for theta in thetas:
    dir = '/storage1/thomasen/hnRNPA1/hnRNPA1_PWrescaling_3/hnRNPA1_PWrescaling_3_productionrun/theta' + str(theta) + '/stage_minusstart'
    chi2r = np.genfromtxt('%s/summary_0_file.dat' % dir, skip_header=1, usecols=(2))
    chi2rlist_withtag.append(chi2r)
```

```python
#get chi2r from summary files
thetas = ['1.00', '1.04', '1.06', '1.07', '1.08', '1.09', '1.10', '1.12']
thetas_float_notag = [1.00, 1.04, 1.06, 1.07, 1.08, 1.09, 1.10, 1.12]

chi2rlist_notag = []
    
for theta in thetas:
    dir = '/lindorffgrp-isilon/thomasen/hnRNPA1/PWrescaling_notag_productionrun_150mM/theta' + str(theta) + '/stage_minusstart'
    chi2r = np.genfromtxt('%s/summary_0_file.dat' % dir, skip_header=1, usecols=(2))
    chi2rlist_notag.append(chi2r)
```

```python
#plot    
fig=plt.figure(figsize=(6,5))
axes=fig.add_axes([1,1,1,1])
    
plt.plot(thetas_float_withtag, chi2rlist_withtag, 'o-', linewidth=2.5, alpha=0.8, markersize=12, label='hSUMO-hnRNPA1*')
plt.plot(thetas_float_notag, chi2rlist_notag, 'o-', linewidth=2.5, alpha=0.8, markersize=12, label='hnRNPA1*')

plt.legend(fontsize=20)

plt.xlabel('λ', fontsize=25)
plt.ylabel('χ$^{2}$$_{r}$', fontsize=25)

plt.axvline(x=1.07, color='gray', linewidth=5, alpha=0.5)
    
plt.xticks(fontsize=20)
plt.yticks(fontsize=20) 

plt.tick_params(colors='black', direction='out', length=7)
    
fig.savefig('/storage1/thomasen/hnRNPA1/PWrescaling_notag_productionrun_150mM/figures/chi2r_vs_lambda_withandwithouttag.pdf', bbox_inches='tight', pad_inches=1)
```

# Rg

```python
#Load tag Rgs from file
thetas = ['1.0', '1.02', '1.04', '1.05', '1.06', '1.07', '1.08', '1.10']
thetas_float_withtag = [1.00, 1.02, 1.04, 1.05, 1.06, 1.07, 1.08, 1.10]
Rg_withtag = []
Rg_avg_withtag = []

directory='/storage1/thomasen/hnRNPA1/hnRNPA1_PWrescaling_3/data/'

#Calculate Rg average from .xvg files and append to Rg_avg
for i in range(len(thetas)):
    Rg_withtag.append(np.genfromtxt(directory + 'Rg_minusstart_hnRNPA1_theta%s.xvg' % thetas[i],skip_header=27,skip_footer=0,usecols=[1],unpack=True))    
    print(len(Rg_withtag[i]))
    Rg_avg_withtag.append(np.average(Rg_withtag[i]))
```

```python
#Load notag Rgs from file
thetas = ['1.00', '1.04', '1.06', '1.07', '1.08', '1.09', '1.10', '1.12']
thetas_float_notag = [1.00, 1.04, 1.06, 1.07, 1.08, 1.09, 1.10, 1.12]

Rg_notag = []
Rg_avg_notag = []

directory='/storage1/thomasen/hnRNPA1/PWrescaling_notag_productionrun_150mM/data/'

#Calculate Rg average from .xvg files and append to Rg_avg
for i in range(len(thetas)):
    Rg_notag.append(np.genfromtxt(directory + 'Rg_gyrate_theta%s.xvg' % thetas[i],skip_header=27,skip_footer=0,usecols=[1],unpack=True))    
    print(len(Rg_notag[i]))
    Rg_avg_notag.append(np.average(Rg_notag[i]))
```

```python
#Plot Rg averages
fig= plt.figure(figsize=(6,5))
axes= fig.add_axes([1,1,1,1])

plt.plot(thetas_float_withtag, Rg_avg_withtag, 'o-', linewidth=2.5, markersize=12, label='hSUMO-hnRNPA1*')
plt.plot(thetas_float_notag, Rg_avg_notag, 'o-', linewidth=2.5, markersize=12, label='hnRNPA1*')

plt.legend(fontsize=20)

#plt.ylim(2.8, 3.3)
#plt.xlim(40, 510)
plt.axvline(x=1.07, color='gray', linewidth=5, alpha=0.5)

plt.xlabel('λ', fontsize=25)
plt.ylabel('R$_g$ (nm)', fontsize=25)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20) 

plt.tick_params(colors='black', direction='out', length=7)

fig.savefig('/lindorffgrp-isilon/thomasen/hnRNPA1/PWrescaling_notag_productionrun_150mM/figures/Rg_vs_lambda_withandwithouttag.pdf', bbox_inches='tight', pad_inches=1)
```
