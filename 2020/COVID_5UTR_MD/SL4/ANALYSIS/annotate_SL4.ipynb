```python
import barnaba as bb
import mdtraj as md

trj1 = md.load("RUN_1/0/SL4_REST_1.crop.xtc",top="sample_SL4.pdb",stride=5)
trj2 = md.load("RUN_2/0/SL4_REST_2.crop.xtc",top="sample_SL4.pdb",stride=5)
```

```python
stack1,pair1,res1 = bb.annotate_traj(trj1)
```

```python
import numpy as np
wc_pairs1 = np.zeros((len(res1),len(res1)))
gu_pairs1 = np.zeros((len(res1),len(res1)))
o_pairs1 = np.zeros((len(res1),len(res1)))
cstack1 = np.zeros((len(res1),len(res1)))
stats1 = {}

for j in range(len(trj1)):
    for k in range(len(pair1[j][0])):
        if(pair1[j][1][k]=="XXX"): continue
        r1 = pair1[j][0][k][0]
        r2 = pair1[j][0][k][1]
        if(pair1[j][1][k] =="WCc" or pair1[j][1][k]=="WWc"):
            wc_pairs1[r1,r2] += 1 
        elif(pair1[j][1][k] =="GUc"):
            gu_pairs1[r1,r2] += 1 
        else:
            o_pairs1[r1,r2] += 1 
            if(pair1[j][0][k][0]==20 and pair1[j][0][k][1]==23):
                if(pair1[j][1][k] in stats1):
                    stats1[pair1[j][1][k]] += 100./len(pair1)
                else:
                    stats1[pair1[j][1][k]] = 100./len(pair1)
    for l in range(len(stack1[j][0])):
        #print(stack1[j][0][l][0],stack1[j][0][l][1])
        r1 = stack1[j][0][l][0]
        r2 = stack1[j][0][l][1]
        cstack1[r1,r2] += 1
print(stats1)

o_pairs1 /= len(trj1)
wc_pairs1 /= len(trj1)
gu_pairs1 /= len(trj1)
cstack1 = 100*cstack1/len(trj1)

for j in range(len(res1)):
    for k in range(j+1,len(res1)):
        ss = 100*wc_pairs1[j,k]+100*gu_pairs1[j,k]+100*o_pairs1[j,k]
        if(ss>2): 
            print("%s %s %5.2f %5.2f %5.2f " % (res1[j],res1[k],100*wc_pairs1[j,k],100*gu_pairs1[j,k],100*o_pairs1[j,k]))
        
        #if(cstack1[j,k]>1):
        #    print("%s %s %5.2f %s " % (res1[j],res1[k],cstack1[j,k],"STACK"))
```
